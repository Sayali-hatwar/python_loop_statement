```python
l2 = list(range(1,100))

previousval = 0
for  l2value in l2:
    #print(l2value)
    currentval = previousval + l2value
    previousval = currentval
    print(f"Previous value : {previousval}, Current value : {currentval}, l2value : {l2value}")
    print()
```
