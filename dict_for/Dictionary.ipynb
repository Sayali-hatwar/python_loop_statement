#### - Dictionary is curly bracket contains key-value pair

```python
d1 = {}

print(f"d1 : {d1}, length of dictionary = {len(d1)}, type of dictionary = {type(d1)}")
```

```python
d2 = {
    "Username": "ARC",
    "Email": "arc@gmail.com",
    "Batch": 25,
    "topic_done":['Excel','SQL','PowerBI']
}

print(f"d2 = {d2}, length of d2 = {len(d2)}, type of d2 = {type(d2)}")
```

```python
print(f"d2['Username']={d2['Username']}, type of d2['Username]= {type(d2['Username'])}")
```
